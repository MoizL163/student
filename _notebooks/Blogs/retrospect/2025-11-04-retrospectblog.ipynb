    const allOptions = ['START', 'STOP', 'CONTINUE', 'MORE', 'LESS'];
    let availableOptions = [...allOptions];
    let spinning = false;

    btn.addEventListener('click', () => {
        if (spinning || availableOptions.length === 0) return;
        spinning = true;
        btn.disabled = true;
        result.textContent = '';

        const randomIndex = Math.floor(Math.random() * availableOptions.length);
        const selected = availableOptions[randomIndex];
        
        availableOptions.splice(randomIndex, 1);
        
        const optionIndex = allOptions.indexOf(selected);
        const extraSpins = 4;
        const totalRotation = (360 * extraSpins) + (optionIndex * 72) + 36;
        
        wheel.style.transform = `rotate(${totalRotation}deg)`;

        setTimeout(() => {
            result.textContent = selected;
            
            if (availableOptions.length === 0) {
                message.textContent = 'All options have been selected!';
                btn.disabled = true;
            } else {
                message.textContent = `${availableOptions.length} option(s) remaining`;
                btn.disabled = false;
            }
            
            spinning = false;
        }, 2000);
    });
</script>

## Q11 — Spinner Simulation

- **My answer:** **B**  
- **Correct answer:** **D**

## Why my answer was wrong
I chose option B because I thought using `RANDOM(1, 3)` would simulate the three sections correctly. But that actually makes each color — Red, Blue, and Yellow — equally likely, each with a 1/3 chance. The problem said Red’s section was four times as large as the others, so my code didn’t reflect that difference in probability.

## Why the correct answer is right
Option D uses `RANDOM(1, 6)` and assigns:
- 1 → Yellow  
- 2 → Blue  
- Everything else (3–6) → Red  

This setup gives Red a 4/6 chance and Blue/Yellow each a 1/6 chance. That matches the description perfectly since Red is supposed to be four times as likely as either of the other colors.

 ## Q14 — Comparing Loop Algorithms

- **My answer:** **B**  
- **Correct answer:** **C**

## Why my answer was wrong
I chose option B because I thought both programs displayed the same numbers but just in a different order. I didn’t look closely enough at *when* each program updates the variable `i` versus when it displays it. I assumed moving the `DISPLAY` and `i ← i + 1` lines wouldn’t change which values appear, but it actually changes the numbers themselves.

## Why the correct answer is right
In **Program A**, the display happens *before* increasing `i`.  
- It starts at `i = 1`, displays 1, then adds 1 each loop → displays **1 through 10**.  

In **Program B**, the increase happens *before* the display.  
- It starts at `i = 1`, immediately adds 1 to make `i = 2`, then displays → displays **2 through 11**.  

So both programs display the **same number of values** (10 total), but the **values differ** — Program A shows 1–10 while Program B shows 2–11.
