**Tools Sprint:** I learned how to use GitHub, Replit, and Jupyter notebooks, which felt confusing at first, but now it’s second nature. I understand how to push commits, write clean markdown, and organize my code. Its so weird how I found something so frustrating when I can do it in 30 seconds now.

**West Coast Travel Sprint:** This one was my favorite. My group focused on a road trip from San Diego to Seattle, visiting stadiums along the way. I worked on the Seattle section — researching sports culture there and finding ways to connect code to real-world data. It felt like an actual project, not just a classroom task.

<script>
    const wheel = document.getElementById('wheel');
    const btn = document.getElementById('spinBtn');
    const result = document.getElementById('result');
    const message = document.getElementById('message');
    const allOptions = ['START', 'STOP', 'CONTINUE', 'MORE', 'LESS'];
    let availableOptions = [...allOptions];
    let spinning = false;

    btn.addEventListener('click', () => {
        if (spinning || availableOptions.length === 0) return;
        spinning = true;
        btn.disabled = true;
        result.textContent = '';

        const randomIndex = Math.floor(Math.random() * availableOptions.length);
        const selected = availableOptions[randomIndex];
        
        availableOptions.splice(randomIndex, 1);
        
        const optionIndex = allOptions.indexOf(selected);
        const extraSpins = 4;
        const totalRotation = (360 * extraSpins) + (optionIndex * 72) + 36;
        
        wheel.style.transform = `rotate(${totalRotation}deg)`;

        setTimeout(() => {
            result.textContent = selected;
            
            if (availableOptions.length === 0) {
                message.textContent = 'All options have been selected!';
                btn.disabled = true;
            } else {
                message.textContent = `${availableOptions.length} option(s) remaining`;
                btn.disabled = false;
            }
            
            spinning = false;
        }, 2000);
    });
</script>
