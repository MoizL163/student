## End of Trimester Reflection

### Compare Yourself to the Beginning of the Year
At the start of the year, I came into the class with the midnset of playing chess and goofing off. I didn’t really connect that mindset to coding. But now, I see how similar they are. Coding is all about logic and predicting outcomes. I went from playing chess on a board to “playing” it in a code editor.

---

### Remember Key Things from 3 Sprints
**Tools Sprint:** I learned how to use GitHub, Replit, and Jupyter notebooks, which felt confusing at first, but now it’s second nature. I understand how to push commits, write clean markdown, and organize my code.

**Fundamentals of JavaScript/Python:** This was the real shift; learning loops, functions, conditionals, and variables. Once I got how they work together, I started seeing patterns like I do in chess. Everything started making more sense because I wasn’t just memorizing; I was actually building things that did something.

**West Coast Travel Sprint:** This one was my favorite. My group focused on a road trip from San Diego to Seattle, visiting stadiums along the way. I worked on the Seattle section — researching sports culture there and finding ways to connect code to real-world data. It felt like an actual project, not just a classroom task.

---

### How N@tM Went
Night at the Museum went way better than I expected. A bunch of people were actually interested in our “Stadium Quest” idea, traveling up the coast and coding it out. Some liked the visuals and said it felt like a real travel app. One person mentioned they’d love to see live game data or team stats for each stadium, which gave me ideas for adding an API later. It was cool hearing real feedback instead of just a grade, like seeing people interact with something I helped build.

### What I Want to Do Next on the Project
I want to expand the Seattle section with more interactive features, maybe buttons that show fun facts, upcoming games, or even weather for the stadium. If I get more time, I want to make a “route planner” that uses distance data between stadiums.  
Eventually, I’d like to connect a sports API to pull live scores or team data automatically. That would make it feel like a legit app, not just a demo.
